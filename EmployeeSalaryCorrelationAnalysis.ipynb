<h2> Correlation Analysis of Employee Data </h2>

<p>For this project I decided to apply a correlation analysis between the Salary of a fictious company and the years of experience of each employee from a dataset found on Kaggle, https://www.kaggle.com/datasets/mohithsairamreddy/salary-data. The goal was to practice analytical concepts to determine if the years of experience are positively correlated to the salary an employee earns. </p>

```python
import pandas as pd
import seaborn as sns
```

```python
df_employees = pd.read_csv(r"C:\Projects\Datasets\Employee_Data.csv")
df_employees
```

<h3> Cleaning the Dataset </h3>
<p> We can see imediately that the elements in the Education Level column are inconsistent so we will rename the elements to maintain consistency </p>

```python
#Reassigning the names of the education level credentials to be more consistent, i.e. renaming Bachelor's to Bachelor's Degree etc.
df_employees["Education Level"].value_counts() #Apply the value_counts function to see what elements are inconsistent
```

```python
#Apply the replace function to change the elements in the column
df_employees["Education Level"] = df_employees["Education Level"].replace({'Bachelor\'s' : 'Bachelor\'s Degree', 'Master\'s' : 'Master\'s Degree', 'phD' : 'PhD'})
df_employees["Education Level"].value_counts()
```

```python
#Apply the same method to check other non-numerical valued columns
df_employees["Gender"].value_counts() #The gender column has no issues and does not need any changes
```

<h4> Correlation Analysis of the Data </h4>
<p> Now we can apply a linear regression to determine any correlation between the Years of Experience of an Employee and their Salary at this fictious company</p>

```python
z = sns.regplot(x = "Years of Experience", y = "Salary", data=df_employees)
z.set_title("Years of Experience vs. Salary")
z.set_xlabel("Years of Experience")
```

<h5> Coffecient of Correlation </h5>
<p>We can clearly see that the line of best fit of the plot shows there is a positive correlation bewteen the Years of Experience and the Salary of the employees. We can also use the Coffecient of Correlation to further show this. </p>

```python
#Calculate the correlation coeffiect with the pandas corr() function
df_employees["Years of Experience"].corr(df_employees["Salary"])
```

<p> Thus we can see that we have a very strong positive correlation as the value of the correlation coefficient is between 0.5 and 1. </p>

<p>We can take it further and see how the relation is amongst the different Education Levels of the employees in the company.</p>

```python
sns.lmplot(x="Years of Experience", y="Salary", hue="Education Level", data=df_employees)
```

We can see visually that there is a positive correlation between the Years of Experience and Salary regardless of the Education Level of the employee.
